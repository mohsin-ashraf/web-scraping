```python
#just to handle some wierd charactors
#Read This article for more details : https://wiki.python.org/moin/UnicodeEncodeError
def safeStr(obj):
    try: return str(obj)
    except UnicodeEncodeError:
        return obj.encode('ascii', 'ignore').decode('ascii')
    except: return ""
```

```python
#Setting up the script
import mechanize
import lxml.html
import urlparse
import csv
```

```python
br = mechanize.Browser()#br short for browser
br.set_handle_robots(False)
br.addheaders = [("user-agent","Chrome")]
```

```python
url = "https://www.amazon.com/All-New-Amazon-Echo-Dot-Add-Alexa-To-Any-Room/product-reviews/B01DFKC2SO/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
html = br.open(url)
```

```python
html = html.read()
```
